```python
import open3d as o3d
import numpy as np
from scipy.spatial import ConvexHull

def calculate_iou_3d(box1, box2):
    """
    Calculate IoU between 3D bounding boxes
    Args:
        box1, box2: (8, 3) array of vertices coordinates
    Returns:
        iou: 3D IoU value
    """
    # Get min/max bounds for each box
    min1, max1 = get_box_bounds(box1)
    min2, max2 = get_box_bounds(box2)

    # Check if boxes overlap
    if not check_overlap(min1, max1, min2, max2):
        return 0.0

    # Calculate box volumes
    vol1 = calculate_volume(min1, max1)
    vol2 = calculate_volume(min2, max2)

    # Calculate intersection volume
    intersection_vol = calculate_intersection_volume(min1, max1, min2, max2)

    # Calculate IoU
    union_vol = vol1 + vol2 - intersection_vol
    iou = intersection_vol / union_vol if union_vol > 0 else 0.0

    return iou

def get_box_bounds(vertices):
    """Get min/max bounds of box vertices"""
    min_bound = np.min(vertices, axis=0)
    max_bound = np.max(vertices, axis=0)
    return min_bound, max_bound

def check_overlap(min1, max1, min2, max2):
    """Check if two boxes overlap"""
    return np.all(max1 >= min2) and np.all(max2 >= min1)

def calculate_volume(min_bound, max_bound):
    """Calculate volume of box"""
    dimensions = max_bound - min_bound
    return np.prod(dimensions)

def calculate_intersection_volume(min1, max1, min2, max2):
    """Calculate volume of intersection between two boxes"""
    # Get intersection bounds
    intersection_min = np.maximum(min1, min2)
    intersection_max = np.minimum(max1, max2)

    # Check if intersection exists
    if np.any(intersection_max <= intersection_min):
        return 0.0

    # Calculate intersection volume
    intersection_dims = intersection_max - intersection_min
    return np.prod(intersection_dims)

# Test case 1: Identical boxes
iou = calculate_iou_3d(box1, box1)  # Should return 1.0
assert iou == 1.0, f"Expected IoU of 1.0, but got {iou}"

# Test case 2: Non-overlapping boxes
box2 = box1 + np.array([2,2,2])
iou = calculate_iou_3d(box1, box2)  # Should return 0.0
assert iou == 0.0, f"Expected IoU of 0.0, but got {iou}"

# Test case 3: Partially overlapping boxes
box3 = box1 + np.array([0.5, 0.5, 0.5])
iou = calculate_iou_3d(box1, box3)  # Should return a value between 0 and 1
assert 0 < iou < 1, f"Expected IoU between 0 and 1, but got {iou}"

# Test case 4: One box inside another
box4 = box1 * 0.5
iou = calculate_iou_3d(box1, box4)  # Should return the volume ratio of the smaller box to the larger box
expected_iou = calculate_volume(*get_box_bounds(box4)) / calculate_volume(*get_box_bounds(box1))
assert iou == expected_iou, f"Expected IoU of {expected_iou}, but got {iou}"

# Test case 5: Edge touching boxes
box5 = box1 + np.array([1, 0, 0])
iou = calculate_iou_3d(box1, box5)  # Should return 0.0 as they only touch at the edge
assert iou == 0.0, f"Expected IoU of 0.0, but got {iou}"
```

```python
from scipy.spatial.transform import Rotation

def get_box_parameters(vertices):
    """
    Extract box parameters (center, dimensions, orientation) using PCA
    Args:
        vertices: (8, 3) array of box vertices
    Returns:
        center: (3,) array
        dimensions: (3,) array
        rotation_matrix: (3, 3) array
    """
    # Calculate center
    center = np.mean(vertices, axis=0)
    centered = vertices - center

    # Get orientation using PCA
    covariance = np.cov(centered.T)
    eigenvalues, eigenvectors = np.linalg.eigh(covariance)

    # Sort by eigenvalues in descending order
    idx = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[idx]
    rotation_matrix = eigenvectors[:, idx]

    # Ensure right-handed coordinate system
    if np.linalg.det(rotation_matrix) < 0:
        rotation_matrix[:, 2] *= -1

    # Transform points to aligned coordinate system
    aligned = np.dot(centered, rotation_matrix)

    # Calculate dimensions
    min_bound = np.min(aligned, axis=0)
    max_bound = np.max(aligned, axis=0)
    dimensions = max_bound - min_bound

    return center, dimensions, rotation_matrix

def merge_boxes(vertices1, vertices2, weight_by_volume=True):
    """Merge boxes while preserving orientation"""
    # Get parameters of both boxes
    center1, dims1, axes1 = get_box_parameters(vertices1)
    center2, dims2, axes2 = get_box_parameters(vertices2)

    # Use orientation of larger box
    vol1 = np.prod(dims1)
    vol2 = np.prod(dims2)

    main_axes = axes1 if vol1 > vol2 else axes2

    # Transform vertices to aligned coordinate system
    combined = np.vstack([vertices1, vertices2])
    centered = combined - np.mean(combined, axis=0)
    aligned = np.dot(centered, main_axes)

    # Calculate new dimensions
    min_bound = np.min(aligned, axis=0)
    max_bound = np.max(aligned, axis=0)
    new_dims = max_bound - min_bound

    # Generate new box vertices
    template = np.array([
        [-0.5, -0.5, -0.5],
        [0.5, -0.5, -0.5],
        [0.5, 0.5, -0.5],
        [-0.5, 0.5, -0.5],
        [-0.5, -0.5, 0.5],
        [0.5, -0.5, 0.5],
        [0.5, 0.5, 0.5],
        [-0.5, 0.5, 0.5]
    ])

    # Scale template by dimensions
    scaled = template * new_dims

    # Rotate back to world space
    rotated = np.dot(scaled, main_axes.T)

    # Translate to final position
    new_center = np.mean(combined, axis=0)
    new_vertices = rotated + new_center

    return new_vertices

def filter_overlapping_boxes(objects, iou_threshold=0.5, distance_threshold=0.2):
    """Filter and merge overlapping boxes"""
    if not objects:
        return []

    filtered_objects = []
    used = set()

    for i, obj1 in enumerate(objects):
        if i in used:
            continue

        current_group = [obj1]
        vertices1 = obj1['vertices']
        center1 = np.mean(vertices1, axis=0)

        # Find all boxes that overlap with the current box
        for j, obj2 in enumerate(objects[i+1:], i+1):
            if j in used:
                continue

            vertices2 = obj2['vertices']
            center2 = np.mean(vertices2, axis=0)

            # Calculate center distance
            center_distance = np.linalg.norm(center1 - center2)

            # Calculate IoU
            iou = calculate_iou_3d(vertices1, vertices2)

            print(f"Center distance: {center_distance}, IoU: {iou}")

            # If boxes are close enough or overlap significantly
            if center_distance < distance_threshold or iou > iou_threshold:
                current_group.append(obj2)
                used.add(j)

        # Merge all boxes in the current group
        if len(current_group) > 1:
            merged_vertices = current_group[0]['vertices']
            for obj in current_group[1:]:
                merged_vertices = merge_boxes(merged_vertices, obj['vertices'])

            # Create new merged object
            merged_obj = {
                'vertices': merged_vertices,
                'color': current_group[0]['color'],  # Keep color of first object
                'faces': current_group[0]['faces'],  # Keep face structure
                'num_merged': len(current_group)  # Track number of merged boxes
            }
            filtered_objects.append(merged_obj)
        else:
            filtered_objects.append(current_group[0])

        used.add(i)

    return filtered_objects

# Test case 1: No overlapping boxes
objects = [{'vertices': box1, 'color': [1, 0, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box2, 'color': [0, 1, 0], 'faces': [], 'num_merged': 1}]
filtered = filter_overlapping_boxes(objects)
assert len(filtered) == 2, f"Expected 2 boxes, but got {len(filtered)}"

# Test case 2: Partially overlapping boxes
objects = [{'vertices': box1, 'color': [1, 0, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box3, 'color': [0, 1, 0], 'faces': [], 'num_merged': 1}]
filtered = filter_overlapping_boxes(objects)
assert len(filtered) == 1, f"Expected 1 box, but got {len(filtered)}"

# Test case 3: One box inside another
objects = [{'vertices': box1, 'color': [1, 0, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box4, 'color': [0, 1, 0], 'faces': [], 'num_merged': 1}]
filtered = filter_overlapping_boxes(objects)
assert len(filtered) == 1, f"Expected 1 box, but got {len(filtered)}"

# Test case 4: Edge touching boxes
objects = [{'vertices': box1, 'color': [1, 0, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box5, 'color': [0, 1, 0], 'faces': [], 'num_merged': 1}]
filtered = filter_overlapping_boxes(objects)
assert len(filtered) == 2, f"Expected 2 boxes, but got {len(filtered)}"

# Test case 5: Overlapping boxes with different orientations
objects = [{'vertices': box1, 'color': [1, 0, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box6, 'color': [0, 1, 0], 'faces': [], 'num_merged': 1}]
filtered = filter_overlapping_boxes(objects)
assert len(filtered) == 1, f"Expected 1 box, but got {len(filtered)}"

# Test case 6: Multiple overlapping boxes
objects = [{'vertices': box1, 'color': [1, 0, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box3, 'color': [0, 1, 0], 'faces': [], 'num_merged': 1},
            {'vertices': box7, 'color': [0, 0, 1], 'faces': [], 'num_merged': 1}]
filtered = filter_overlapping_boxes(objects)
assert len(filtered) == 1, f"Expected 1 box, but got {len(filtered)}"
```

```python
def create_box_line_set(obj, color):
    """Create LineSet for box edges"""
    vertices = obj['vertices']
    min_bound = np.min(vertices, axis=0)
    max_bound = np.max(vertices, axis=0)

    points = [
        [min_bound[0], min_bound[1], min_bound[2]],
        [max_bound[0], min_bound[1], min_bound[2]],
        [max_bound[0], max_bound[1], min_bound[2]],
        [min_bound[0], max_bound[1], min_bound[2]],
        [min_bound[0], min_bound[1], max_bound[2]],
        [max_bound[0], min_bound[1], max_bound[2]],
        [max_bound[0], max_bound[1], max_bound[2]],
        [min_bound[0], max_bound[1], max_bound[2]]
    ]

    lines = [
        [0, 1], [1, 2], [2, 3], [3, 0],
        [4, 5], [5, 6], [6, 7], [7, 4],
        [0, 4], [1, 5], [2, 6], [3, 7]
    ]

    box_color = obj['color'] if obj['color'] is not None else color

    line_set = o3d.geometry.LineSet()
    line_set.points = o3d.utility.Vector3dVector(points)
    line_set.lines = o3d.utility.Vector2iVector(lines)
    line_set.colors = o3d.utility.Vector3dVector([box_color] * len(lines))

    return line_set

def parse_obj_file(obj_file, iou_threshold=0.5, distance_threshold=0.3):
    """
    Parse obj file and return objects with vertices and colors.
    Each 3D box consists of exactly 8 vertices and 6 faces.
    """
    objects = []
    vertices = []
    faces = []
    current_color = None
    vertex_offset = 0  # Track global vertex index offset

    with open(obj_file, 'r') as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()

        # Parse vertex line
        if line.startswith('v '):
            parts = line.split()
            vertex = [float(x) for x in parts[1:4]]

            # Handle color information
            if len(parts) >= 7:
                color = [int(x)/255 for x in parts[4:7]]
                if not current_color:
                    current_color = color

            vertices.append(vertex)

            # Check if we have collected 8 vertices
            if len(vertices) == 8:
                # Look ahead for 6 faces
                face_count = 0
                face_lines = []
                j = i + 1

                while j < len(lines) and face_count < 6:
                    next_line = lines[j].strip()
                    if next_line.startswith('f '):
                        # Adjust vertex indices based on current offset
                        face = [int(x) - 1 - vertex_offset for x in next_line.split()[1:]]
                        face_lines.append(face)
                        face_count += 1
                    j += 1

                # If we found exactly 6 faces, create a new object
                if face_count == 6:
                    objects.append({
                        'vertices': np.array(vertices),
                        'faces': np.array(face_lines),
                        'color': current_color
                    })

                    # Reset for next object
                    vertices = []
                    current_color = None
                    vertex_offset += 8  # Update offset for next object
                    i = j - 1  # Move iterator to after last face
                else:
                    # If we didn't find 6 faces, something is wrong
                    print(f"Warning: Found 8 vertices but not 6 faces at vertex {vertex_offset + 1}")
                    vertices = vertices[1:]  # Remove first vertex and continue

        i += 1

    print(f"Successfully parsed {len(objects)} objects")
    print(f"Total vertices processed: {vertex_offset}")

    # Validation
    for idx, obj in enumerate(objects):
        if obj['vertices'].shape != (8, 3):
            print(f"Warning: Object {idx} has {obj['vertices'].shape[0]} vertices instead of 8")
        if obj['faces'].shape != (6, 4):
            print(f"Warning: Object {idx} has {obj['faces'].shape[0]} faces instead of 6")

    # Filter and merge overlapping boxes
    objects = filter_overlapping_boxes(objects, iou_threshold, distance_threshold)

    return objects

def visualize_objects(obj_file, point_cloud_file=None, iou_threshold=0.5, distance_threshold=0.3):
    """Visualize point cloud with filtered and merged bounding boxes"""
    # Parse objects with improved parser
    objects = parse_obj_file(obj_file, iou_threshold, distance_threshold)
    print(f"Found {len(objects)} objects")

    colors = assign_unique_colors(len(objects))


    geometries = []

    # Add point cloud if available
    if point_cloud_file:
        try:
            pcd = o3d.io.read_point_cloud(point_cloud_file)
            geometries.append(pcd)
        except Exception as e:
            print(f"Error loading point cloud: {e}")

    # Create and add bounding boxes
    for obj, fallback_color in zip(objects, colors):
        line_set = create_box_line_set(obj, fallback_color)
        geometries.append(line_set)

        # Print debug information
        vertices = obj['vertices']
        min_bound = np.min(vertices, axis=0)
        max_bound = np.max(vertices, axis=0)
        dimensions = max_bound - min_bound
        center = (max_bound + min_bound) / 2
        color = obj['color'] if obj['color'] is not None else fallback_color

        num_merged = obj.get('num_merged', 1)
        print(f"Box dimensions: {dimensions}")
        print(f"Box center: {center}")
        print(f"Box color: {color}")
        if num_merged > 1:
            print(f"Merged from {num_merged} boxes")

    # Visualize
    if geometries:
        o3d.visualization.draw_geometries(geometries,
                                        window_name="Point Cloud with Merged Bounding Boxes",
                                        mesh_show_back_face=True)
    else:
        print("No geometries to visualize")

child = "scene0382_00"
path = f"/home/quantran2381/unidet3d/work_dirs/unidet3d_1xb8_scannet/scannet/{child}/{child}_"

to_ply(path + "points.obj", path + "points.ply", 'obj')

visualize_objects(
    obj_file=path + "pred.obj",
    point_cloud_file=path + "points.ply"
)
```
